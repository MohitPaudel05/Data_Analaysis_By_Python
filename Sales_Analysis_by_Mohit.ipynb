**Exploratory Data Analysis**

**1. Sales by Gender**





```python
ax = sns.countplot( data =df, x='Gender',hue='Gender')
for x in ax.containers:
    ax.bar_label(x)
```

```python
sales_by_amt= df.groupby(['Gender'])['Amount'].sum().sort_values(ascending=False)
sns.barplot(x= 'Gender',y ='Amount',data = sales_by_amt.reset_index(), hue='Gender')
plt.show()
```

**From the above Graph, we can see that the female buyers are more**



**2. Sales According to Age**

```python
df['Age Group'].nunique()
```

```python
ax = sns.countplot(x='Age Group',data= df, hue="Gender")
for x in ax.containers:
    ax.bar_label(x)
```

```python
sales_by_amt= df.groupby(['Age Group','Gender'])['Amount'].sum().sort_values(ascending = False)
sns.barplot(x='Age Group',y="Amount",data = sales_by_amt.reset_index(),hue='Gender')
plt.show()
```

```python
sales_by_order= df.groupby(['Age Group','Gender'],as_index = False)['Orders'].sum().sort_values(by ="Orders",ascending = False)
ax =sns.barplot(x= "Age Group", y="Orders",data = sales_by_order, hue="Gender",palette='Set1' )
for x in ax.containers:
    ax.bar_label(x)
```

 **From the above graph, we can see maximum buyers are of Female age group 26-35 with an order of appox 8217**

**3.  Sales by Marital_status**

```python
ax= sns.countplot(x= 'Marital_Status', data = df,hue='Gender', palette='Set1')
for x in ax.containers:
    ax.bar_label(x)
```

**3. Sales By State**

```python
df.columns
```

```python
df['State'].nunique()
```

```python
sales_of_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(8)
sns.set_theme(rc={'figure.figsize':(15,4)})
ax= sns.barplot(x='State',y ='Orders', data = sales_of_state, hue= 'State')
for x in ax.containers:
    ax.bar_label(x)
    
```

```python
orders_of_state = df.groupby(['State', 'Gender'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).tail(8)
sns.set_theme(rc={'figure.figsize':(15,4)})
ax= sns.barplot(x='State',y ='Orders', data = orders_of_state, hue= 'State')
for x in ax.containers:
    ax.bar_label(x)
```

```python
sales_of_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(7)
sns.set_theme(rc={'figure.figsize':(15,4)})
ax= sns.barplot(x='State',y ='Amount', data = sales_of_state, hue= "State", )
```

**From above graphs we can see that most of the orders and total sales amount are from Uttar Pradesh, Maharashtra and Karnataka respectively**

***4. Sales By Zone***

```python
df.columns
```

```python
orders_of_state= df.groupby(['Zone'],as_index= False)['Orders'].sum().sort_values(by="Orders",ascending = False)
ax= sns.barplot(x ="Zone", y= "Orders", data = orders_of_state, hue = 'Zone')
for x in ax.containers:
    ax.bar_label(x)
```

**From above graphs we can see that most of the orders and total sales amount are from Central, Southern and Western respectively**

**5. Sales By Occupation***

```python
df['Occupation'].nunique()
```

```python
ax = sns.countplot(data = df,x ='Occupation',hue = 'Occupation')
sns.set(rc ={'figure.figsize':(22,6)})
for x in ax.containers:
    ax.bar_label(x)
```

```python
sales_of_occ= df.groupby(['Occupation'],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending =False)
sns.barplot(x ='Occupation', y='Amount', data = sales_of_occ, hue = "Occupation")
sns.set(rc ={'figure.figsize':(22,6)})
```

**From above graphs we can see that most of the orders and total sales amount are from IT sector, Healthcare and Aviation respectively**

***6. Maximum Sales of Product_Category***

```python
df['Product_Category'].nunique()
```

```python
# plt.figure(dpi=200)
sns.set(rc ={'figure.figsize':(23,6)})
ax= sns.countplot(data = df, x="Product_Category")
for x in ax.containers:
    ax.bar_label(x)



```

```python
sales_of_product= df.groupby(['Product_Category'], as_index = False)["Amount"].sum().sort_values(by ="Amount", ascending= False).head(12)
sns.set(rc ={'figure.figsize':(23,6)})

sns.barplot(x ='Product_Category',y="Amount",data = sales_of_product, hue = 'Product_Category')
```

**From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category**



***7. Maximum Sale of ProductID***

```python
sales_of_productID = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,4)})
sns.barplot(data = sales_of_productID, x = 'Product_ID',y= 'Orders', hue="Product_ID")
```
